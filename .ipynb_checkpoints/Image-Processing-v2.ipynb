A hands-on walkthrough of common image operations using the **Pillow** library.

| Section | Operation |
|---------|-----------|
| 1 | Setup & Load |
| 2 | Save |
| 3 | Crop |
| 4 | Rotate & Flip |

## 1 · Setup & Load

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# ── helper: display one or more images side-by-side ──────────────────────────
def show(*images, titles=None, cmap=None):
    """Display one or more PIL Images in a single row."""
    n = len(images)
    fig, axes = plt.subplots(1, n, figsize=(6 * n, 5))
    if n == 1:
        axes = [axes]
    for ax, img, title in zip(axes, images, titles or [None] * n):
        ax.imshow(img, cmap=cmap)
        ax.axis("off")
        if title:
            ax.set_title(title, fontsize=13)
    plt.tight_layout()
    plt.show()
```

```python
PHOTO = "redbull.jpg"

img = Image.open(PHOTO)
print(f"Size  : {img.size[0]} × {img.size[1]} px")
print(f"Mode  : {img.mode}")
print(f"Format: {img.format}")

show(img, titles=["Original"])
```

## 2 · Save

```python
OUTPUT_PATH = "saved.jpg"
img.save(OUTPUT_PATH)
print(f"Saved to '{OUTPUT_PATH}'")

# Verify by reloading
img_saved = Image.open(OUTPUT_PATH)
show(img_saved, titles=["Reloaded from disk"])
```

## 3 · Crop
`Image.crop(box)` expects a 4-tuple **(left, upper, right, lower)** in pixel coordinates.


```python
# Crop a region from the original image
box = (100, 100, 700, 450)          # (left, upper, right, lower)
img_crop = img.crop(box)

print(f"Original : {img.size}")
print(f"Cropped  : {img_crop.size}")
show(img, img_crop, titles=["Original", f"Cropped {box}"])
```

## 4 · Rotate & Flip
- `rotate(angle)` — counter-clockwise rotation; use `expand=True` to avoid clipping.  
- `transpose(method)` — lossless 90 ° / 180 ° / flip operations.


```python
img_r45  = img.rotate(45, expand=True)
img_r90  = img.transpose(Image.ROTATE_90)
img_flip = img.transpose(Image.FLIP_LEFT_RIGHT)

show(img_r45, img_r90, img_flip,
     titles=["rotate(45)", "ROTATE_90", "FLIP_LEFT_RIGHT"])
```

## 5 · Resize
Pass a `(width, height)` tuple.  Use `Image.LANCZOS` for the best downscaling quality.


```python
TARGET = (400, 300)
img_small = img.resize(TARGET, resample=Image.LANCZOS)

print(f"Original : {img.size}")
print(f"Resized  : {img_small.size}")
show(img, img_small, titles=["Original", f"Resized to {TARGET}"])
```

## 6 · Colour Conversions
| Mode | Description |
|------|-------------|
| `RGB` | Standard colour (default) |
| `L` | Greyscale (0 – 255) |
| `1` | Binary black-and-white |
| `F` | 32-bit float (for scientific use) |


```python
img_grey = img.convert("L")

# Binary: threshold each pixel
THRESHOLD = 128
img_binary = img_grey.point(lambda px: 255 if px > THRESHOLD else 0).convert("1")

show(img, img_grey, img_binary,
     titles=["RGB", "Greyscale (L)", f"Binary (threshold={THRESHOLD})"],
     cmap="gray")
```

## 7 · Filters
PIL's `ImageFilter` module provides a set of built-in convolution kernels.

| Filter | Effect |
|--------|--------|
| `BLUR` | Smooths the image |
| `SHARPEN` | Enhances detail |
| `EDGE_ENHANCE` | Boosts edges slightly |
| `EMBOSS` | Embossed / relief look |
| `CONTOUR` | Extracts outlines |


```python
filters = {
    "BLUR"        : ImageFilter.BLUR,
    "SHARPEN"     : ImageFilter.SHARPEN,
    "EDGE_ENHANCE": ImageFilter.EDGE_ENHANCE,
    "EMBOSS"      : ImageFilter.EMBOSS,
    "CONTOUR"     : ImageFilter.CONTOUR,
}

fig, axes = plt.subplots(1, len(filters), figsize=(5 * len(filters), 5))
for ax, (name, filt) in zip(axes, filters.items()):
    ax.imshow(img.filter(filt))
    ax.set_title(name, fontsize=12)
    ax.axis("off")
plt.suptitle("PIL Filter Comparison", fontsize=15, y=1.02)
plt.tight_layout()
plt.show()
```

## Summary

| Operation | Key method |
|-----------|-----------|
| Open / save | `Image.open()` · `img.save()` |
| Inspect | `img.size` · `img.mode` · `img.format` |
| Crop | `img.crop((left, upper, right, lower))` |
| Rotate | `img.rotate(angle, expand=True)` |
| Flip / transpose | `img.transpose(Image.ROTATE_90 / FLIP_*)` |
| Resize | `img.resize((w, h), resample=Image.LANCZOS)` |
| Convert colour | `img.convert("L" / "1" / "RGB")` |
| Pixel ops | `img.point(func)` |
| Filters | `img.filter(ImageFilter.*)` |
