### Cosine Similarity

```python
cosine_sim = cosine_similarity(tfidf_matrix)
cosine_sim
```

```python
# For each movie, get average similarity to its top 10 most similar movies
def get_avg_similarity(cosine_sim, n=10):
    scores = [] # empty lis tto store each movie's final score
    for i in range(len(cosine_sim)): # go through every movie
        row = cosine_sim[i].copy()  # grab movie i's full row — its similarity to ALL other movies
        row[i] = 0  # exclude self
        top_k = np.sort(row)[::-1][:n]  # top k scores
        scores.append(top_k.mean())
    return scores
imdb['avg_cosine_score'] = get_avg_similarity(cosine_sim, n=10)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# Distribution of avg cosine scores
axes[0].hist(imdb['avg_cosine_score'], bins=50, color='steelblue', edgecolor='white')
axes[0].axvline(imdb['avg_cosine_score'].mean(), color='red', linestyle='--', label=f"Mean: {imdb['avg_cosine_score'].mean():.3f}")
axes[0].set_title("Distribution of Avg Cosine Similarity Scores")
axes[0].set_xlabel("Cosine Score")
axes[0].set_ylabel("Number of Movies")
axes[0].legend()
# Content rating breakdown
rating_counts = imdb['content_rating'].value_counts().sort_index()
axes[1].bar(rating_counts.index, rating_counts.values, color=sns.color_palette("crest", 5))
axes[1].set_title("Movies per Content Rating Tier")
axes[1].set_xlabel("Content Rating (1=Low, 5=High Similarity)")
axes[1].set_ylabel("Number of Movies")
for i, v in enumerate(rating_counts.values):
    axes[1].text(rating_counts.index[i], v + 5, str(v), ha='center', fontsize=11)
plt.tight_layout()
plt.show()
```

```python
def cosine_to_rating(score):
    if score < 0.25:   return 1   
    elif score < 0.33: return 2   
    elif score < 0.37: return 3   
    elif score < 0.42: return 4   
    else:              return 5   # top performers

imdb['content_rating'] = imdb['avg_cosine_score'].apply(cosine_to_rating)
```

```python
imdb[['movie_title', 'imdb_score', 'avg_cosine_score', 'content_rating']].sort_values('imdb_score', ascending=False).head(20)
```

### Testing Cosine Similarity find Recommendations

```python
def get_recommendations(title, top_n=10):
    
    if title not in imdb['movie_title'].values:
        print(f" '{title}' not found.")
        return None

    # get movie index
    idx = imdb[imdb['movie_title'] == title].index[0]
    
    # get similarity scores
    sim_scores = cosine_sim[idx]
    
    # get top_n most similar movies (excluding itself)
    top_indices = np.argsort(sim_scores)[::-1][1:top_n+1]
    top_scores  = sim_scores[top_indices]

    # build results dataframe
    results = imdb[['movie_title', 'genres_str', 'director_name', 'imdb_score']].iloc[top_indices].copy()
    results['cosine_score']   = top_scores.round(4)
    results['content_rating'] = results['cosine_score'].apply(cosine_to_rating)

    print(f"\n🎬 Recommendations for: '{title}'\n")
    return results.reset_index(drop=True)
#
get_recommendations("Avatar")
```

Since this is a content-based recommender system, we evaluate it using:
- **Intra-list similarity** — are recommended movies similar to each other?
- **Cosine score distribution** — how well does the model separate similar vs dissimilar movies?
- **Genre hit rate** — do recommendations share genres with the query movie?
- **IMDB score comparison** — are recommended movies rated comparably to the query?

### 1 · Genre Hit Rate
Measures how often recommended movies share at least one genre with the query movie.

```python
def genre_hit_rate(title, top_n=10):
    """% of recommended movies sharing at least one genre with the query."""
    if title not in imdb['movie_title'].values:
        print(f"'{title}' not found.")
        return None
    idx = imdb[imdb['movie_title'] == title].index[0] #index of Avatar movie 
    # genre related to Avatar Movie
    query_genres = set(imdb.loc[idx, 'genres']) if isinstance(imdb.loc[idx, 'genres'], list) else set()
    # Similarity Score of movies similar to Avatar 
    sim_scores  = cosine_sim[idx]
    # Get top 10 movie indices which are similar to Avatar 
    top_indices = np.argsort(sim_scores)[::-1][1:top_n+1]
    hits = 0
    for i in top_indices:
        rec_genres = set(imdb.loc[i, 'genres']) if isinstance(imdb.loc[i, 'genres'], list) else set()
        if query_genres & rec_genres:
            hits += 1
    rate = hits / top_n
    print(f"Query       : {title}")
    print(f"Genres      : {query_genres}")
    print(f"Genre Hits  : {hits}/{top_n}")
    print(f"Hit Rate    : {rate:.0%}")
    return rate
```

```python
# Test on multiple movies
test_movies = ["Avatar", "The Dark Knight", "Inception", "Titanic", "The Lion King"]
hit_rates   = {}
for movie in test_movies:
    if movie in imdb['movie_title'].values:
        rate = genre_hit_rate(movie)
        hit_rates[movie] = rate
        print()

print(f"\nAverage Genre Hit Rate: {np.mean(list(hit_rates.values())):.0%}")
```

### 2 · IMDB Score Comparison
Compares the average IMDB score of recommended movies vs the query movie.

```python
def score_comparison(title, feature, top_n=10):
    if title not in imdb['movie_title'].values:
        return None
    idx = imdb[imdb['movie_title'] == title].index[0]
    query_score = imdb.loc[idx, feature]        

    sim_scores  = cosine_sim[idx]
    top_indices = np.argsort(sim_scores)[::-1][1:top_n+1]
    rec_scores  = imdb.iloc[top_indices][feature]
    return query_score, rec_scores

def print_values(query_score, rec_scores):
    print(f"Query Movie Score : {query_score:.1f}")
    print(f"Avg Recommended Score: {rec_scores.mean():.2f}")
    print(f"Min Recommended Score: {rec_scores.min():.1f}")
    print(f"Max Recommended Score: {rec_scores.max():.1f}")
    print(f"Score Difference : {abs(query_score - rec_scores.mean()):.2f}")

print_values(*score_comparison("Avatar", 'imdb_score'))
```

We evaluate the dataset using 4 ML models across 2 tasks:

| Task | Target | Models |

```python
# Use numeric features + avg_cosine_score as ML features
feature_cols = ['duration', 'num_voted_users', 'num_user_for_reviews', 'avg_cosine_score']
ml_df = imdb[feature_cols + ['imdb_score', 'content_rating']].dropna()

X = ml_df[feature_cols]
y_reg  = ml_df['imdb_score']       # regression target
y_clf  = ml_df['content_rating']   # classification target
# Scale features
scaler  = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, yr_train, yr_test, yc_train, yc_test = train_test_split(X_scaled, y_reg, y_clf, test_size=0.2, random_state=42)
print(f"Train size: {X_train.shape[0]}")
print(f"Test size: {X_test.shape[0]}")
print(f"Features: {feature_cols}")
```

### 📈 Task 1 · Regression — Predict IMDB Score

####  RMSE — Root Mean Squared Error

Measures how far off predictions are from actual values.

$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

**Example:**
| Actual | Predicted | Error |
|--------|-----------|-------|
| 7.9 | 7.5 | 0.4 |
| 6.5 | 6.8 | 0.3 |
| 8.2 | 7.9 | 0.3 |

$$RMSE = \sqrt{\frac{0.4^2 + 0.3^2 + 0.3^2}{3}} = 0.34$$

- **Lower is better** — 0 means perfect predictions
- Same unit as the target variable (e.g. IMDB score)
- RMSE of 0.34 means predictions are off by ~0.34 points on average

---

#### R² — R Squared (Coefficient of Determination)

Measures how well the model explains the variance in the data.

$$R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}$$

| R² Value | Meaning |
|----------|---------|
| 0.9+ | Excellent |
| 0.7 – 0.9 | Good |
| 0.5 – 0.7 | Moderate |
| < 0.5 | Poor |
| 0 | No better than predicting the mean |
| < 0 | Worse than predicting the mean |

- **Higher is better** — 1.0 means perfect fit
- Tells you how much of the pattern the model captured



```python
reg_models = {
    'Linear Regression' : LinearRegression(),
    'KNN Regressor'     : KNeighborsRegressor(n_neighbors=5),
    'Random Forest'     : RandomForestRegressor(n_estimators=100, random_state=42),
}
reg_results = {}
for name, model in reg_models.items():
    model.fit(X_train, yr_train)
    preds = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(yr_test, preds))
    r2 = r2_score(yr_test, preds)
    reg_results[name] = {'RMSE': round(rmse, 4), 'R² Score': round(r2, 4)}
    print(f"{name:25} RMSE: {rmse:.4f}  R²: {r2:.4f}")

reg_df = pd.DataFrame(reg_results).T
print("\n", reg_df)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Regression Results — Predicting IMDB Score", fontsize=14, weight='bold')
# RMSE comparison
sns.barplot(x=reg_df.index, y=reg_df['RMSE'], palette='crest', ax=axes[0])
axes[0].set_title("RMSE (lower is better)")
axes[0].set_ylabel("RMSE")
axes[0].tick_params(axis='x', rotation=15)
for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%.4f', padding=4)
# R² comparison
sns.barplot(x=reg_df.index, y=reg_df['R² Score'], palette='flare', ax=axes[1])
axes[1].set_title("R² Score (higher is better)")
axes[1].set_ylabel("R²")
axes[1].tick_params(axis='x', rotation=15)
for container in axes[1].containers:
    axes[1].bar_label(container, fmt='%.4f', padding=4)
plt.tight_layout()
plt.show()
```

## 🏷️ Task 2 · Classification — Predict Content Rating (1–5)


```python
def model_evaluation(model_name, y_test, y_pred):
    cm     = confusion_matrix(y_test, y_pred)
    labels = [str(i) for i in sorted(y_test.unique())]

    report    = classification_report(y_test, y_pred, output_dict=True)
    report_df = pd.DataFrame(report).T.round(3)
    accuracy  = report_df.loc['accuracy', 'precision']
    report_df = report_df.drop(index='accuracy')
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle(f"Model Evaluation — {model_name}", fontsize=14, weight='bold')
    # Left — confusion matrix
    sns.heatmap(cm, annot=True, fmt='d', cmap='RdYlGn',xticklabels=labels, yticklabels=labels, ax=axes[0])
    axes[0].set_title("Confusion Matrix")
    axes[0].set_xlabel("Predicted")
    axes[0].set_ylabel("Actual")
    # Right — classification report as heatmap
    report_vals = report_df[['precision', 'recall', 'f1-score']]
    sns.heatmap(report_vals, annot=True, fmt='.2f', cmap='RdYlGn',ax=axes[1], vmin=0, vmax=1)
    axes[1].set_title(f"Classification Report: Accuracy: {accuracy:.2%}")
    plt.tight_layout()
    plt.show()


for name, model in clf_models.items():
    model.fit(X_train, yc_train)
    preds = model.predict(X_test)
    model_evaluation(name, yc_test, preds)
```

  meaning it misses some class 1 movies
- Classes **2–5** all score above 0.90 across all metrics
- Some confusion between **adjacent classes** (e.g. 2→3, 3→4) which is
  expected since neighbouring rating tiers are naturally similar

### Random Forest — Accuracy: 99.70%
- Near perfect performance across all 5 classes

- Only **3 misclassifications in class 5** out of 194 samples
- All precision, recall and f1-scores at **1.00** except class 5 recall (0.98)

## 🔵 Task 3 · K-Means Clustering
Group movies by content features — no target variable needed.


#### Using Silhouette Score to Mathematically Predict Clusters

```python
silhouette_scores = []
K_range = range(2, 11)
for k in K_range:
    km  = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    score  = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)
    print(f"K={k}  Silhouette Score: {score:.4f}")
plt.figure(figsize=(8, 4))
plt.plot(K_range, silhouette_scores, marker='o', color='steelblue', linewidth=2)
plt.title("Silhouette Score — Optimal Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.xticks(K_range)
plt.tight_layout()
plt.show()

best_k = K_range[silhouette_scores.index(max(silhouette_scores))]
print(f"\nOptimal K: {best_k}  (Score: {max(silhouette_scores):.4f})")
```

<div style="background-color: #f0f4f8; padding: 15px; border-radius: 8px; border-left: 5px solid #3498db;">

#### Observation: Silhouette Score — Optimal K

#### Results
| K | Silhouette Score |
|---|-----------------|
| **2** | **0.5546** ← highest |
| 3 | 0.2986 |
| 4 | 0.3084 |
| 5 | 0.2957 |
| 6 | 0.2434 |
| 7 | 0.2546 |
| 8 | 0.2440 |
| 9 | 0.2357 |
| 10 | 0.2416 |

#### Analysis
- **K=2** has the highest silhouette score of **0.5546** — mathematically optimal
- Sharp drop from K=2 to K=3, after which scores remain low and flat
- Scores below 0.3 (K=3 onwards) indicate weak cluster separation

#### Key Insight
> Although the silhouette score identifies **K=2** as mathematically optimal,
> this produces only 2 broad groups which is **too coarse** for a rich movie dataset. So I will use elbow method to find clusters instead.

#### Using Elbow Method

```python
# Elbow method — find optimal K
wcss= []
K_range = range(2, 11)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    wcss.append(km.inertia_)
plt.figure(figsize=(8, 4))
plt.plot(K_range, wcss, marker='o', color='steelblue', linewidth=2)
plt.title("Elbow Method — Optimal Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.xticks(K_range)
plt.tight_layout()
plt.show()
```

```python
km = KMeans(n_clusters=4, random_state=42, n_init=10)
ml_df = ml_df.copy()
ml_df['cluster'] = km.fit_predict(X_scaled)
# Visualise with PCA
pca   = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
plt.figure(figsize=(9, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1],c=ml_df['cluster'], cmap='tab10', alpha=0.5, s=15)
plt.colorbar(scatter, label='Cluster')
plt.title("K-Means Clusters visualised with PCA (2D)")
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]:.0%} variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]:.0%} variance)")
plt.tight_layout()
plt.show()
```

<div style="background-color: #f0f4f8; padding: 15px; border-radius: 8px; border-left: 5px solid #3498db;">

#### Observation: Elbow Method — K-Means Clustering

#### Why K=4?
While the silhouette score suggests K=2 as optimal, the elbow method
supports **K=4** as a more meaningful choice for this dataset:

- The steepest drop in inertia occurs between **K=2 → K=4**
- Beyond K=4, the curve begins to flatten — diminishing returns set in
- K=2 produces only 2 broad groups which is **too coarse** for a movie dataset
  with diverse content, engagement levels, and rating tiers

#### PCA Visualisation Support
The PCA plot confirms K=4 is a reasonable choice:
- **Red cluster** — dense, tightly packed → low engagement movies
- **Blue cluster** — spread vertically → moderate engagement, varied content
- **Teal cluster** — transitional group between low and high engagement
- **Pink cluster** — spread horizontally along PC1 → high voted, popular movies

PC1 explains **52% of variance** and PC2 explains **24%** — together
capturing **76% of total variance**, making the 2D visualisation reliable.

#### Key Insight
> K=2 is mathematically optimal but **practically too simple**.
> K=4 strikes the right balance — capturing distinct movie segments
> (low engagement, moderate, high engagement, and outlier blockbusters)
> while avoiding overfitting the cluster structure.
