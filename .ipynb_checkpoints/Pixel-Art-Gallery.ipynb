# 🎨 Pixel Art Gallery
All images built from scratch using numpy arrays and PIL.

```python
from PIL import Image, ImageFilter
import numpy as np
import matplotlib.pyplot as plt
```

## 🚦 Traffic Light

```python
tl = np.zeros((60, 25, 3), dtype=np.uint8)
tl[:, :] = [30, 30, 30]

for i in range(60):
    for j in range(25):
        if (i-10)**2 + (j-12)**2 <= 36:
            tl[i, j] = [255, 0, 0]
        if (i-30)**2 + (j-12)**2 <= 36:
            tl[i, j] = [255, 220, 0]
        if (i-50)**2 + (j-12)**2 <= 36:
            tl[i, j] = [0, 200, 0]

img_tl = Image.fromarray(tl).resize((250, 600), Image.NEAREST)
plt.imshow(img_tl)
plt.title("Traffic Light 🚦")
plt.axis("off")
plt.show()
```

## 🇺🇸 American Flag

```python
flag = np.zeros((40, 80, 3), dtype=np.uint8)

stripe_h = 40 // 13
for i in range(13):
    color = [220, 0, 0] if i % 2 == 0 else [255, 255, 255]
    flag[i*stripe_h:(i+1)*stripe_h, :] = color

flag[:18, :30] = [0, 40, 150]

star_positions = [(2,3),(2,9),(2,15),(2,21),(2,27),
                  (5,6),(5,12),(5,18),(5,24),
                  (8,3),(8,9),(8,15),(8,21),(8,27),
                  (11,6),(11,12),(11,18),(11,24),
                  (14,3),(14,9),(14,15),(14,21),(14,27)]
for (r, c) in star_positions:
    flag[r:r+2, c:c+2] = [255, 255, 255]

img_flag = Image.fromarray(flag).resize((600, 300), Image.NEAREST)
plt.imshow(img_flag)
plt.title("American Flag 🇺🇸")
plt.axis("off")
plt.show()
```

## 🌅 Sunset

```python
sunset = np.zeros((40, 80, 3), dtype=np.uint8)

colors = [
    [20,  20,  80],
    [80,  40,  120],
    [180, 80,  60],
    [220, 120, 40],
    [240, 180, 60],
]
band = 40 // len(colors)
for i, color in enumerate(colors):
    sunset[i*band:(i+1)*band, :] = color

for i in range(40):
    for j in range(80):
        if (i-28)**2 + (j-40)**2 <= 25:
            sunset[i, j] = [255, 240, 50]

sunset[32:, :] = [20, 60, 100]

for i in range(40):
    for j in range(80):
        if (i-36)**2 + (j-40)**2 <= 12:
            sunset[i, j] = [255, 200, 50]

img_sunset = Image.fromarray(sunset).resize((600, 300), Image.NEAREST)
plt.imshow(img_sunset)
plt.title("Sunset 🌅")
plt.axis("off")
plt.show()
```

## 😊 Happy Face

```python
face = np.ones((20, 20, 3), dtype=np.uint8) * 255

for i in range(20):
    for j in range(20):
        if (i-10)**2 + (j-10)**2 <= 81:
            face[i, j] = [255, 220, 0]

face[6:8,  6:8]  = [0, 0, 0]
face[6:8, 12:14] = [0, 0, 0]

for (r, c) in [(13,6),(13,7),(14,8),(14,9),(14,10),(14,11),(13,12),(13,13)]:
    face[r, c] = [0, 0, 0]

img_face = Image.fromarray(face).resize((400, 400), Image.NEAREST)
plt.imshow(img_face)
plt.title("Happy Face 😊")
plt.axis("off")
plt.show()
```

## 🏎️ Red Bull F1 Car

```python
car = np.zeros((20, 40, 3), dtype=np.uint8)
car[:, :] = [255, 255, 255]

car[8:15, 5:35]  = [0, 0, 139]
car[8:10, 5:35]  = [220, 0, 0]
car[13:15, 5:35] = [220, 0, 0]
car[4:9, 12:28]  = [0, 0, 120]
car[5:9, 13:27]  = [20, 20, 60]
car[11:13, 1:7]  = [0, 0, 139]
car[12:13, 1:4]  = [220, 0, 0]
car[14:16, 1:12] = [0, 0, 100]
car[15:16, 1:14] = [220, 0, 0]
car[5:9,  33:38] = [0, 0, 100]
car[5:6,  33:38] = [220, 0, 0]
car[8:15, 5:7]   = [255, 200, 0]
car[8:15, 33:35] = [255, 200, 0]
car[10:13, 19:22]= [255, 200, 0]
car[14:19, 8:14] = [20, 20, 20]
car[15:18, 9:13] = [255, 200, 0]
car[14:19, 26:32]= [20, 20, 20]
car[15:18, 27:31]= [255, 200, 0]
car[11:13, 33:36]= [255, 100, 0]

img_car = Image.fromarray(car).resize((800, 400), Image.NEAREST)
plt.imshow(img_car)
plt.title("Red Bull F1 Car 🏎️")
plt.axis("off")
plt.show()
```

## 🏁 F1 Chequered Flag

```python
flag_f1 = np.zeros((50, 80, 3), dtype=np.uint8)
flag_f1[:, :] = [135, 206, 235]
flag_f1[:, 2:6] = [80, 80, 80]

square = 5
for row in range(8):
    for col in range(10):
        color = [255, 255, 255] if (row + col) % 2 == 0 else [0, 0, 0]
        r0, r1 = 5 + row*square, 5 + (row+1)*square
        c0, c1 = 6 + col*square, 6 + (col+1)*square
        flag_f1[r0:r1, c0:c1] = color

img_f1 = Image.fromarray(flag_f1).resize((640, 400), Image.NEAREST)
plt.imshow(img_f1)
plt.title("F1 Chequered Flag 🏁")
plt.axis("off")
plt.show()
```

## 🌳 Pixel Landscape

```python
land = np.zeros((20, 20, 3), dtype=np.uint8)
land[:10, :]     = [135, 206, 235]
land[10:, :]     = [34, 139, 34]
land[1:4, 15:18] = [255, 255, 0]
land[12:18, 9:11]= [139, 69, 19]
land[8:13, 7:13] = [0, 100, 0]

img_land = Image.fromarray(land).resize((400, 400), Image.NEAREST)
plt.imshow(img_land)
plt.title("Pixel Landscape 🌳")
plt.axis("off")
plt.show()
```

## 🟥 4-Color Grid

```python
grid = np.zeros((20, 20, 3), dtype=np.uint8)
n = 20
r = n // 2
grid[:r, :r] = [255, 0,   0  ]
grid[:r, r:] = [0,   255, 0  ]
grid[r:, :r] = [0,   0,   255]
grid[r:, r:] = [255, 255, 0  ]

img_grid = Image.fromarray(grid).resize((400, 400), Image.NEAREST)
plt.imshow(img_grid)
plt.title("4-Color Grid 🟥🟩🟦🟨")
plt.axis("off")
plt.show()
```
