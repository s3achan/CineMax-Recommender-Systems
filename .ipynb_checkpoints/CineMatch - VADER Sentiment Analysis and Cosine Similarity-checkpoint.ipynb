<h3>🎭 VADER Sentiment Analysis</h3>
</div>

<div style="background-color:#fff9c4; padding:16px; border-radius:8px; border-left:5px solid #fbc02d;">
VADER (Valence Aware Dictionary and sEntiment Reasoner) is a lexicon and rule-based sentiment analysis tool specifically designed for social media and short informal text. It produces a compound score ranging from **-1 (most negative)** to **+1 (most positive)**.

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
rotten_tomatoes ['compound_score'] = rotten_tomatoes ['review_content'].apply(
    lambda x: analyzer.polarity_scores(str(x))['compound'] if pd.notna(x) else None
)
```

```python
pd.set_option('future.no_silent_downcasting', True)
# Average sentiment per movie
movie_sentiment = (rotten_tomatoes.groupby('movie_title')['compound_score'].mean().reset_index().rename(columns={'compound_score': 'sentiment_score'}))
print("Movies with sentiment scores:", len(movie_sentiment))
movie_sentiment.head(5)
```

### Distribution of Sentiment Score

```python
plt.figure(figsize=(8,4))
ss = movie_sentiment['sentiment_score'].dropna()
plt.hist(ss, bins=50, color='steelblue', edgecolor='white')
plt.title('Distribution of Sentiment Score')
plt.xlabel('Sentiment Score')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
movie_sentiment['sentiment_score'].describe()
```

### Bucketing Sentiment Score into 1–5 Rating

```python
rotten_tomatoes = rotten_tomatoes.drop(columns=['sentiment_score', 'sentiment_rating'], errors='ignore')
```

```python
# -1.0 ←—[1]—→ -0.50 ←—[2]—→ 0.0 ←—[3]—→ 0.30 ←—[4]—→ 0.65 ←—[5]—→ 1.0
def vader_to_rating(score):
    if pd.isna(score):  return None   # unknown → neutral
    elif score < -0.50: return 1   # very negative  (~bottom 15%)
    elif score < 0.00:  return 2   # mostly negative
    elif score < 0.30:  return 3   # mixed/neutral
    elif score < 0.65:  return 4   # mostly positive
    else:               return 5   # very positive  (~top 25%)

movie_sentiment['sentiment_rating'] = movie_sentiment['sentiment_score'].apply(vader_to_rating)
rotten_tomatoes = rotten_tomatoes.merge(movie_sentiment, on='movie_title', how='left')
# Check results
rotten_tomatoes[['movie_title', 'sentiment_score', 'sentiment_rating']].dropna().head(5)
```

<h3>🔗 Content-Based Filtering — Cosine Similarity</h3>

### Feature Engineering & TF-IDF Vectorizer

```python
# Clean actors and directors
rotten_tomatoes['actors_str']  = rotten_tomatoes['actors'].fillna('').str.replace(',', ' ')
rotten_tomatoes['directors']   = rotten_tomatoes['directors'].fillna('')
# Build features column: genres + directors + actors
rotten_tomatoes['features'] = (rotten_tomatoes['genres_str'] + " " +rotten_tomatoes['directors']  + " " +rotten_tomatoes['actors_str'])
rotten_tomatoes['features'].head(3)
```

```python
# Deduplicate on movie_title before TF-IDF
rotten_tomatoes = rotten_tomatoes.drop_duplicates(subset='movie_title').reset_index(drop=True)
print("After dedup:", rotten_tomatoes.shape)
```

```python
tfidf        = TfidfVectorizer(stop_words='english', max_features=5000, min_df=2, ngram_range=(1, 2))
tfidf_matrix = tfidf.fit_transform(rotten_tomatoes['features'])
print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
```

### Average Cosine Similarity Score

```python
# For each movie, get average similarity to its top 10 most similar movies
def get_avg_similarity(cosine_sim, n=10):
    scores = [] # empty lis tto store each movie's final score
    for i in range(len(cosine_sim)): # go through every movie
        row = cosine_sim[i].copy()  # grab movie i's full row — its similarity to ALL other movies
        row[i] = 0  # exclude self
        top_k = np.sort(row)[::-1][:n]  # top k scores
        scores.append(top_k.mean())
    return scores
rotten_tomatoes['avg_cosine_score'] = get_avg_similarity(cosine_sim, n=10)
```

### Distribution of Avg Cosine Scores

```python
plt.figure(figsize=(8,4))
plt.hist(rotten_tomatoes['avg_cosine_score'], bins=50, color='steelblue', edgecolor='white')
plt.title('Distribution of Avg Cosine Scores')
plt.xlabel('Cosine Score')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
print(rotten_tomatoes['avg_cosine_score'].describe())
```

### Bucketing Cosine Score into 1–5 Rating

```python
# Thresholds based on actual percentiles from distribution
# 0.0 ←—[1]—→ 0.19 ←—[2]—→ 0.23 ←—[3]—→ 0.31 ←—[4]—→ 0.45 ←—[5]—→ 1.0
def cosine_to_rating(score):
    if score < 0.19:   return 1   
    elif score < 0.23: return 2   
    elif score < 0.31: return 3   
    elif score < 0.45: return 4   
    else:              return 5   # top performers

rotten_tomatoes['cosine_rating'] = rotten_tomatoes['avg_cosine_score'].apply(cosine_to_rating)
```

```python
rotten_tomatoes[['movie_title', 'avg_cosine_score', 'cosine_rating']].head(5)
```

<div style="border-bottom: 3px solid #64748b; padding: 8px 0; margin: 25px 0;">
<h3>🎬 Hybrid Recommender</h3>
</div

<div style="background-color:#fff9c4; padding:16px; border-radius:8px; border-left:5px solid #fbc02d;">

The hybrid recommender combines:
- **Content Score** (`cosine_rating`) — how similar a movie is to the input based on genres, directors, and actors
- **Sentiment Score** (`sentiment_rating`) — how positively reviewed the movie is based on VADER analysis
- **Alpha** — controls the weight between content and sentiment (default: 60% content, 40% sentiment)

```python
def get_recommendations(title, top_n=10, alpha=0.6):
    if title not in rotten_tomatoes['movie_title'].values:
        print(f"'{title}' not found.")
        return None

    idx         = rotten_tomatoes[rotten_tomatoes['movie_title'] == title].index[0]
    sim_scores  = cosine_sim[idx]
    top_indices = np.argsort(sim_scores)[::-1][1:top_n+1]
    top_scores  = sim_scores[top_indices]

    results = rotten_tomatoes[['movie_title', 'genres_str', 'directors', 'cosine_rating','sentiment_rating']].iloc[top_indices].copy()
    
    results['cosine_score'] = top_scores.round(4)
    results['hybrid_score'] = (alpha * results['cosine_rating'] + (1 - alpha) * results['sentiment_rating']).round(3)

    results = results.sort_values('hybrid_score', ascending=False)
    print(f"\n🎬 Recommendations for: '{title}'\n")
    return results.reset_index(drop=True)

get_recommendations("The Godfather")
```

<div style="border-bottom: 3px solid #64748b; padding: 8px 0; margin: 25px 0;">
<h3>🎛️ Alpha Sensitivity: Content vs. Sentiment Weighting</h3>
</div

```python
def plot_alpha_comparison(title, alphas=[0.2, 0.4, 0.6, 0.8]):
    if title not in rotten_tomatoes['movie_title'].values:
        print(f" '{title}' not found.")
        return None
        
    idx  = rotten_tomatoes[rotten_tomatoes['movie_title'] == title].index[0]
    sim_scores = cosine_sim[idx]
    top_indices = np.argsort(sim_scores)[::-1][1:11]

    results = rotten_tomatoes[['movie_title', 'cosine_rating', 'sentiment_rating']].iloc[top_indices].copy()
    median_sentiment = rotten_tomatoes['sentiment_rating'].median()
    results['sentiment_rating'] = results['sentiment_rating'].fillna(median_sentiment)

    # Shorten long titles for display
    results['short_title'] = results['movie_title'].apply(lambda x: x[:25] + '...' if len(x) > 25 else x)
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle(f"How Alpha Affects Recommendations for: '{title}'", fontsize=14, weight='bold')
    axes = axes.flatten()

    for i, alpha in enumerate(alphas):
        results['hybrid_score'] = (alpha * results['cosine_rating'] +(1 - alpha) * results['sentiment_rating']).round(3)
        results_sorted = results.sort_values('hybrid_score', ascending=True)
        bars = axes[i].barh(results_sorted['short_title'],results_sorted['hybrid_score'],color=sns.color_palette("pastel")[i],edgecolor='white')
        axes[i].bar_label(bars, fmt='%.2f', padding=4, fontsize=10)
        axes[i].set_title(f"Alpha = {alpha}  |  Content: {int(alpha*100)}%  Sentiment: {int((1-alpha)*100)}%",fontsize=11)
        axes[i].set_xlabel("Hybrid Score")
        axes[i].set_xlim(0, 6)
```

```python
plot_alpha_comparison("The Godfather")
```

<h3>📝 Project Summary</h3>

<h4>🎯 Objective</h4>
This project builds a <strong>Hybrid Movie Recommendation System</strong> that combines 
Content-Based Filtering and VADER Sentiment Analysis to recommend movies based on both 
content similarity and critical reception.

<h4>📂 Dataset</h4>
The Rotten Tomatoes Movies & Critic Reviews dataset from Kaggle was used, consisting of 
over <strong>17,000 unique movies</strong> and <strong>168,000+ critic reviews</strong>. 
The two files were merged on <code>rotten_tomatoes_link</code> to combine movie metadata 
with review text.

<h4>🧹 Data Cleaning & Preprocessing</h4>
<ul>
    <li>Missing categorical values filled with <code>"Unknown"</code></li>
    <li>Movie titles cleaned — removed invisible characters and whitespace</li>
    <li>Genres saved as string before splitting for EDA</li>
    <li>Deduplicated on <code>movie_title</code> before TF-IDF modeling</li>
</ul>

<h4>🎭 VADER Sentiment Analysis</h4>
<ul>
    <li>VADER compound scores computed per review on <code>review_content</code></li>
    <li>Scores averaged per movie to produce a single <code>sentiment_score</code> (-1 to 1)</li>
    <li>Bucketed into a <strong>1–5 sentiment rating</strong> based on actual score distribution</li>
    <li>Movies with missing reviews assigned <code>None</code> — handled with median in recommendations</li>
</ul>

<h4>🔗 Content-Based Filtering</h4>
<ul>
    <li>Features built from <code>genres + directors + actors</code></li>
    <li>TF-IDF Vectorizer applied with <code>max_features=5000</code>, bigrams, <code>min_df=2</code></li>
    <li>Similarity computed using <code>linear_kernel</code> (faster equivalent of cosine similarity)</li>
    <li>Average similarity to top-10 neighbors computed per movie</li>
    <li>Bucketed into a <strong>1–5 cosine rating</strong> based on percentile thresholds</li>
</ul>

<h4>🎬 Hybrid Recommender</h4>
<ul>
    <li>Combines <strong>cosine rating</strong> (content) and <strong>sentiment rating</strong> (reviews)</li>
    <li>Controlled by <strong>alpha</strong> — default 60% content, 40% sentiment</li>
    <li>Alpha sensitivity analysis shows how recommendations shift at different weightings</li>
</ul>

<h4>📊 Key Findings</h4>
<ul>
    <li>Most movies score <strong>low on cosine similarity</strong> (ratings 1–3) — the dataset is diverse and content-unique</li>
    <li>Most movies score <strong>high on sentiment</strong> (ratings 4–5) — critics tend to be positive overall</li>
    <li>High sentiment ≠ high content similarity — combining both produces more balanced recommendations</li>
    <li>Alpha tuning significantly shifts recommendations — lower alpha favours critically acclaimed films regardless of genre</li>
</ul>

<h4>🚀 Future Improvements</h4>
